# ML MODEL SELECTION

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

# CLUSTRING

```python
car.head(2)
```

```python
car['PowerTrain'].replace(to_replace=['RWD','FWD','AWD'],value=[0, 1, 2],inplace=True)
car['RapidCharge'].replace(to_replace=['No','Yes'],value=[0, 1],inplace=True)
```

```python
#feature selecting
X=car[['AccelSec','TopSpeed_KmH','Efficiency_WhKm','FastCharge_KmH', 'Range_Km', 'RapidCharge', 'Seats', 'PriceEuro','PowerTrain']]
X
```

```python
# feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
# applying Principle Component Analysis (PCA)
pca = PCA(n_components=9)
X_pca = pca.fit_transform(X_scaled)
df_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7', 'PC8', 'PC9'])
df_pca.head()
```

```python
# training the model using k=4 as rendered by the above plot
kmean = KMeans(n_clusters=4, init='k-means++', random_state=90)
kmean.fit(X_pca)
```

```python
print(kmean.labels_)
```

```python
# check the size of clusters
pd.Series(kmean.labels_).value_counts()
```

```python
car['clusters'] = kmean.labels_
```

```python
kmean.cluster_centers_
```

```python
# visualizing clusters
plt.figure(figsize=(7,5))
sns.scatterplot(data=df_pca, x='PC1', y='PC9', s=70, hue=kmean.labels_, palette='viridis', zorder=2, alpha=.9)
plt.scatter(x=kmean.cluster_centers_[:,0], y=kmean.cluster_centers_[:,1], marker="*", c="r", s=80, label="centroids")
plt.grid()
plt.show()
```
